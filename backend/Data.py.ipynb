```python
import pandas as pd
from model import ChurnPredictor 

predictor = ChurnPredictor()
my_dataset = predictor.generate_sample_data(n_samples=1000)
my_dataset.to_csv('synthetic_churn_data.csv', index=False)
print("Data successfully generated and saved to synthetic_churn_data.csv!")

my_dataset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# We will use the 'my_dataset' you generated in the previous step.
# (Make sure you ran the previous code block so my_dataset exists in memory!)

# ---------------------------------------------------------
# OPTION 1: BAR CHART (Churn based on Contract Type)
# This shows how month-to-month contracts have higher churn.
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.countplot(data=my_dataset, x='contract_type', hue='churn', palette='Set2')
plt.title('Customer Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.legend(title='Churn (1=Yes, 0=No)')
plt.show()

# ---------------------------------------------------------
# OPTION 2: PIE CHART (Overall Churn Distribution)
# This shows the total percentage of customers who left vs stayed.
# ---------------------------------------------------------
plt.figure(figsize=(6, 6))
churn_counts = my_dataset['churn'].value_counts()
plt.pie(churn_counts, labels=['Retained (0)', 'Churned (1)'], 
        autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'], startangle=90)
plt.title('Overall Customer Churn Distribution')
plt.show()
```
